This exercise relates to the College data set, which can be found in
the file College.csv on the ISLR book website. It contains a number of
variables for 777 different universities and colleges in the US. The
variables are <br>
• Private : Public/private indicator <br>
• Apps : Number of applications received <br>
• Accept : Number of applicants accepted <br>
• Enroll : Number of new students enrolled <br>
• Top10perc : New students from top 10 % of high school class <br>
• Top25perc : New students from top 25 % of high school class <br>
• F.Undergrad : Number of full-time undergraduates <br>
• P.Undergrad : Number of part-time undergraduates <br>
• Outstate : Out-of-state tuition <br>
• Room.Board : Room and board costs <br>
• Books : Estimated book costs <br>
• Personal : Estimated personal spending <br>
• PhD : Percent of faculty with Ph.D.s <br>
• Terminal : Percent of faculty with terminal degree <br>
• S.F.Ratio : Student/faculty ratio <br>
• perc.alumni : Percent of alumni who donate <br>
• Expend : Instructional expenditure per student <br>
• Grad.Rate : Graduation rate <br>
Before reading the data into Python, it can be viewed in Excel or a
text editor.

### 1. Explore the 'Private' Column:

```python
# Count the number of private and public institutions using value_counts() or groupby() and count()
college['College Type'].value_counts()
```

```python
#Visualize the distribution of private and public institutions using pie chart and count chart in two subplots using seaborn
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='College Type', data=college)
plt.title('Count of Private and Public Colleges')
plt.subplot(1, 2, 2)
plt.pie(college['College Type'].value_counts(), labels=['Private', 'Public'], autopct='%.1f%%')
plt.title('Percentage of Private and Public Colleges')
plt.show()



```

```python
#Calculate summary statistics for other variables (e.g., 'Enroll,' 'Grad.Rate') separately for private and public institutions to identify differences.
college.groupby('College Type').agg({'Enroll': ['mean', 'median', 'std'], 'Grad.Rate': ['mean', 'median', 'std']})
```

```python
### Plot the count of applicants and percentage of accepted students according to college type (public/private)

college_type_apps = college.groupby('College Type').agg({'Apps': ['sum'], 'Accept': ['sum']})
college_type_apps.columns = ['Applications', 'Accepted']
college_type_apps.reset_index(inplace=True)
plt.figure(figsize=(10, 10))
ax = sns.barplot(college_type_apps.melt(id_vars=['College Type'], value_name='Number of Applications', var_name='Applications vs Accepted').head(),
            x='College Type', y='Number of Applications', hue='Applications vs Accepted')

#show values on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge')

plt.legend()
# Calculate and plot the percentage of accepted applications in private and public colleges
college_type_apps['Percent Accepted'] = (college_type_apps['Accepted'] / college_type_apps['Applications']) * 100
plt.ylabel('Number of Applications/Accepted')

ax2 = plt.twinx()
sns.lineplot(x=college_type_apps.index, y='Percent Accepted', data=college_type_apps, color='red', marker='o', linestyle='--', label='Percent Accepted', ax=ax2)

# Add labels and legend
plt.ylabel('Percent Accepted')
plt.xlabel('College Type')
plt.title('Applications Count and Acceptance Percentage by College Type')
plt.legend(bbox_to_anchor=(1.0,0.93))
```

```python
# Relation between college type full time and part time undergrads
plt.figure(figsize=(10, 5))
sns.scatterplot(x='F.Undergrad', y='P.Undergrad', data=college, hue='College Type')
plt.title('Full Time vs Part Time Undergrads by College Type')
```

```python
# Relation between college type and total full time and total part time undergrads in barplot 
college_type_undergrads = college.groupby('College Type').agg({'F.Undergrad': ['sum'], 'P.Undergrad': ['sum']})
college_type_undergrads.columns = ['Full Time Undergrads', 'Part Time Undergrads']
college_type_undergrads.reset_index(inplace=True)
plt.figure(figsize=(10, 10))

ax = sns.barplot(x='College Type', y='Full Time Undergrads', data=college_type_undergrads, label='Full Time Undergrads', color='orange', alpha=0.8)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge')
ax = sns.barplot(x='College Type', y='Part Time Undergrads', data=college_type_undergrads, label='Part Time Undergrads', color='skyblue', alpha=0.8)
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', label_type='edge')
plt.legend()
plt.title('Full Time vs Part Time Undergrads by College Type')


```
